```python
C_t = 0.75
I_a = np.arange(0.03, 0.20, 0.005)


k_star = 0.11 * (C_t**1.07) * (I_a**0.20)
epsilon = 0.23 * (C_t**-0.25) * (I_a**0.17)
a = 0.93 * (C_t**-0.75) * (I_a**0.17)
b = 0.42 * (C_t**0.6) * (I_a**0.2)
c = 0.15 * (C_t**-0.25) * (I_a**-0.7)
# d = 2.3 * (C_t**-1.2)
e = 1.0 * (I_a**0.1)
f = 0.7 * (C_t**-3.2) * (I_a**-0.45)
y = [k_star, epsilon, a, b, c, e, f,]
labels = ["k_star", "epsilon", "a", "b", "c", "e", "f",]
line_styles = ["-", "--", "-.", ":", "-", "--", "-."]

fig = plt.figure(dpi=200)
ax = fig.add_subplot(111)
colors = list(mcolors.TABLEAU_COLORS.keys())
for i, label in enumerate(labels):
    ax.plot(I_a, y[i], color=colors[i], linewidth=2, linestyle=line_styles[i], label=label)
ax.legend(loc="best", edgecolor='None', frameon=False,
          labelspacing=0.4, bbox_transform=ax.transAxes)
ax.set_xlabel("Ambient Turbulence Intensity")
ax.set_ylabel("Parameter Value")
plt.show()
```
